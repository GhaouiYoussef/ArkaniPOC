```python
import torch
import librosa
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
import os

# Define the constants
LANG_ID = "ar"
MODEL_ID = "jonatasgrosman/wav2vec2-large-xlsr-53-arabic"
audio_path = "/notebooks/1/Quran_Ayat_public/audio_data/Abdul_Basit_Murattal_64kbps/001001.mp3"

# Load the processor and model
processor = Wav2Vec2Processor.from_pretrained(MODEL_ID)
model = Wav2Vec2ForCTC.from_pretrained(MODEL_ID)

# Load the audio file using librosa
speech_array, sampling_rate = librosa.load(audio_path, sr=16_000)

# Process the audio file with the processor
inputs = processor(speech_array, sampling_rate=16_000, return_tensors="pt", padding=True)

# Pass the inputs through the model
with torch.no_grad():
    logits = model(inputs.input_values, attention_mask=inputs.attention_mask).logits

# Get the predicted ids
predicted_ids = torch.argmax(logits, dim=-1)

# Decode the ids to get the predicted sentence
predicted_sentence = processor.batch_decode(predicted_ids)

# Output the result
predicted_sentence
```

```python

# Function to process a batch of files
def process_audio_batch(dataset):
    predictions = []
    for index, row in dataset.iterrows():
        # Load the audio file using librosa
        speech_array, sampling_rate = librosa.load(row["PATH"], sr=16_000)
        
        # Process the audio file with the processor
        inputs = processor(speech_array, sampling_rate=16_000, return_tensors="pt", padding=True)

        # Pass the inputs through the model
        with torch.no_grad():
            logits = model(inputs.input_values, attention_mask=inputs.attention_mask).logits

        # Get the predicted ids and decode them
        predicted_ids = torch.argmax(logits, dim=-1)
        predicted_sentence = processor.batch_decode(predicted_ids)
        
        # Store the predicted sentence
        predictions.append(predicted_sentence[0])
    
    return predictions

BATCH_SIZE = 10
# Evaluate the batch
random_rows = transcript_df.sample(BATCH_SIZE).reset_index(drop=True)
predicted_sentences = process_audio_batch(random_rows)

# Print the reference and predictions for each audio sample
for i, predicted_sentence in enumerate(predicted_sentences):
    print("-" * 100)
    print("Reference:", random_rows["TRANSCRIPT"][i])
    print("Prediction:", predicted_sentence)
```
